# While loop 



While loop is used to repeat the block of code until it dooes not reach the specified condition 

```python
l1=[1,2,3,4,5]

for i in l1:
    print(i)
```

```python
l1=[1,2,3,4,5]
timer =0

while timer <2:
    for i in l1:
        if i <3:
            print(i)
        timer+=1
```

```python
## user will assign number and it will keep adding that number untill user does not assign 0
user_input=int(input("Enter the input "))

result=0

while user_input!= 0:
    result=result+user_input
    user_input=int(input("Enter the input "))
    print(result)
print(result)
    


```
