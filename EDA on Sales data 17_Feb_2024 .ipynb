```python
## Let's merge fact_sale and dim_product based on inner join 

sale=pd.merge(fact_sale,dim_product,how='inner',left_on='Product Code',right_on='Product code')
```

```python
sale.head()
```

```python
## Let's join sale with dim customer 

sale=sale.merge(dim_customer,how='inner', left_on='Customer _code',right_on='Customer Code')
```

```python
sale.head()
```

```python
## let join the dim_state
sale=sale.merge(dim_state,how='inner',left_on='State code',right_on='State Code')
```

### I have observed that after the merge operation below mentioned columns are twice in the dataframe 

- Product code
- Customer code
- State code

I am removing all these columns from the dataframe

```python
sale.drop(columns=['Customer _code','Product code','State Code'],inplace=True)
```

```python
sale.head()
```

```python
## rename the unit sale price price columns 

sale.rename(columns={'Unit sales price Price':'Unit Sale Price'},inplace=True)
```

```python
sale.head()
```

```python
## Let's chnage all the float value into two decimal points
sale['Sales Amount']=round(sale['Sales Amount'],2)

sale['Unit cost  Price']=round(sale['Unit cost  Price'],2)

sale['Unit Sale Price']=round(sale['Unit Sale Price'],2)
```

```python
# waht is the total sale 
    
print("Total sale including all the year",np.sum(sale['Sales Amount']))
```

```python
## What is the average sale 

print("The average sale including all the year ",round(np.mean(sale['Sales Amount']),2))
```

```python
## sale by category 

sale_by_category=pd.pivot_table(data=sale,index='Category',values='Sales Amount',aggfunc=sum)
```
