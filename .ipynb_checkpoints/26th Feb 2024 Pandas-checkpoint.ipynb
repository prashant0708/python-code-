```python
movies_data
```

```python
movies_data
```

```python
for index, i in movies_data.iterrows():
    if i['unit']=='Billions':
        movies_data.at[index,'Budget_Millions']*=1000
    if i['unit']=='Thousands':
        movies_data.at[index,'Budget_Millions']/=1000
    
```

```python
movies_data['revenue_millions']=movies_data.apply(lambda x :x['revenue']*1000 if x['unit']=='Billions' else x['revenue']/1000 if x['unit']=='Thousands'  else x['revenue'],axis=1)
```

```python
movies_data.groupby(['Industry','Movie_Name'])['Budget_Millions'].sum()
```

```python
pd.pivot_table(movies_data,index=['Industry','Movie_Name'],values=['Budget_Millions','revenue'],aggfunc=sum)
```

```python
### how many movies released in the 2022 year

movies_in_2022=movies_data[movies_data['release_year']==2022]
```
