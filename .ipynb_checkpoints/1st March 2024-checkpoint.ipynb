```python
xls=pd.ExcelFile(r"C:\Users\prash\Downloads\Store data.xlsx")
```

```python
store2=pd.read_excel(xls,sheet_name='Store 2')
```

```python
store1=pd.read_excel(xls,sheet_name='Store 1')
```

```python
store1
```

```python
store1.dtypes
```

```python
## Extracted the year from date

store1['Year']=store1['Date'].dt.year
```

```python
store1
```

```python
### extract the month from the date

store1["Month"]=store1['Date'].dt.month
```

```python
store1["Month"].unique()
```

```python
### extract the month name from the data columns

store1["Month_Name"]=store1["Date"].dt.month_name()
```

```python
store1
```

```python
### extract the day for the date columns

store1['Day']=store1['Date'].dt.day
```

```python
store1
```

```python
store1['Day_name']=store1['Date'].dt.day_name()
```

```python
### Create one columns weekend and weekday and whereever the day is Sunday and Saturday mark as weekend else weekday

store1["Weekday"]=store1['Day_name'].apply(lambda x :  "Weekend" if x=="Sunday" or x=="Saturday" else "weekday")
```

```python
store1
```

```python
### get the sales by week day and weekend 

pd.pivot_table(data=store1,index="Weekday",values='Sales Amount',aggfunc=sum)
```

```python
writer=pd.ExcelWriter(r"C:\Users\prash\Downloads\Result_store1.xlsx")
```

```python
store1.to_excel(writer,sheet_name='Store1',index=False)
store2.to_excel(writer,sheet_name='Store2',index=False)
```
