## Extraction of Day , Month , Year etc from Date columns

```python
Stores.dtypes
```

```python
## Extract the Year from date columns
Stores['Year']=Stores['Date'].dt.year
```

```python
Stores.head(2)
```

```python
## Extract the Month 
Stores['Month']=Stores['Date'].dt.month
```

```python
Stores.head(2)
```

```python
## Extract the days
Stores['day']=Stores['Date'].dt.day
```

```python
Stores.head(2)
```

```python
## Extract the Month name from Date columns
Stores['Month_name']=Stores['Date'].dt.month_name()
```

```python
Stores.head(2)
```

```python
## Extract the frist 3 character

Stores['extracted']=Stores['Department'].str[:3]
```

```python
Stores.head()
```

```python
## Extract the last 3 character
Stores['extracted_last']=Stores['Department'].str[-3:]
```

```python
Stores.head()
```

```python
for index, i in Stores.iterrows():
    if i['Store']=='Store 1':
        Stores.at[index,"Quantity"]= Stores['Quantity']+5
        
```

```python
Stores
```

```python
round(Stores['Sales Amount'].sum(),2)
```

```python
## find the sales amount by Store 

Stores.groupby(['Store'])['Sales Amount'].sum()
```

```python
## Find the Sales by store and department 

Stores.groupby(['Store','Department'])['Sales Amount'].sum()
```

```python
## to save in excel 
result=Stores.groupby(['Store','Department'])['Sales Amount'].sum()
```
